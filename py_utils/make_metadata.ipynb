### Building the metadata table

The script gows through the dataset folder, reads which systems are present and some of their features creating a table with that information

```python
import pandas as pd
import os
```

```python
def read_description(folder):
  try:
    file = open(folder + '/description.txt', 'r')
    description = file.read()
    if description[-1] == '\n':
      description = description[:-1]
    file.close()
  except:
    description = ""
  return description
```

```python
dataset_folder = '/content/ComponentSystemsData/datasets/'

metadata = pd.DataFrame(columns=['label', 'description', 'n_objects', 'n_components', 'total_size'])

for filename in os.listdir(dataset_folder):

  if not os.path.isdir(dataset_folder + filename):
    continue

  i = len(metadata.index)
  metadata.loc[i, 'label'] = filename
  metadata.loc[i, 'description'] = read_description(dataset_folder + filename)

  obj_path = dataset_folder + filename + '/data/objects.tsv'
  if os.path.isfile(obj_path):
    objects = pd.read_csv(obj_path, sep='\t')
    metadata.loc[i, 'n_objects'] = len(objects)
    metadata.loc[i, 'total_size'] = sum(objects['size'])

  comp_path = dataset_folder + filename + '/data/components.tsv'
  if os.path.isfile(comp_path):
    components = pd.read_csv(comp_path, sep='\t')
    metadata.loc[i, 'n_components'] = len(components)
```

```python
metadata.to_csv('metadata.tsv', sep='\t', index=False)
metadata
```
