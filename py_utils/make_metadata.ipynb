Going through the dataset folder and reading which systems are present and their main information.
The result is exported in the metadata table.

```python
dataset_folder = '/content/ComponentSystemsData/datasets/books_gutenberg/'
```

```python
dataset_folder = '/content/ComponentSystemsData/datasets/'

metadata = pd.DataFrame(columns=['label', 'category', 'n_objects', 'n_components', 'total_size', 'description', 'link'])

for filename in os.listdir(dataset_folder):

  if not os.path.isdir(dataset_folder + filename):
    continue

  i = len(metadata.index)
  descr_path = dataset_folder + filename + '/description.tsv'
  #if not os.path.isfile(descr_path):
  #  continue
  descr = pd.read_csv(descr_path, sep='\t', header=None, index_col=0)[1]

  metadata.loc[i, 'label'] = filename
  metadata.loc[i, 'category'] = descr.loc['Category']
  metadata.loc[i, 'description'] = descr.loc['Description']
  metadata.loc[i, 'link'] = descr.loc['Link']

  obj_path = dataset_folder + filename + '/data/objects.tsv'
  if os.path.isfile(obj_path):
    objects = pd.read_csv(obj_path, sep='\t')
    metadata.loc[i, 'n_objects'] = len(objects)
    metadata.loc[i, 'total_size'] = sum(objects['size'])

  comp_path = dataset_folder + filename + '/data/components.tsv'
  if os.path.isfile(comp_path):
    components = pd.read_csv(comp_path, sep='\t')
    metadata.loc[i, 'n_components'] = len(components)
```

```python
metadata.to_csv('metadata.tsv', sep='\t', index=False)
metadata
```
