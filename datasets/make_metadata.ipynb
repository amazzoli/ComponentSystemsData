```python
import pandas as pd
import os
```

```python
def read_description(folder):
  try:
    file = open(folder + '/description.txt', 'r')
    description = file.read()
    file.close()
  except:
    description = ""
  return description
```

```python
dataset_folder = '/content/ComponentSystemsData/datasets/'

labels = []

for filename in os.listdir(dataset_folder):

  if os.path.isdir(dataset_folder + filename):
    labels.append(filename)

  print(read_description(dataset_folder + filename))
```
